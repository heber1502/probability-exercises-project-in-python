## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random 
import numpy as np

intentos = 1000 

# Funcion lanzar dado

def lanzar_dado():
    return random.randint(1,6)

# Contador 
contador = 0

# Simulacion de lanzamiento de dados

for i in range(intentos):
    dado1 = lanzar_dado()
    dado2 = lanzar_dado()
    suma = dado1 + dado2 

    if suma > 7 or suma % 2 == 0:
        contador += 1 

# Calculando probabilidad estimada
probabilidad_estimada = contador / intentos

print(f"Probabilidad estimada de que la suma sea mayor a 7 o par: {np.round(probabilidad_estimada*100, 2)}%")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
import random
import numpy as np

# caja
caja = ["White"] * 10 + ["Red"] * 20 + ["Green"] * 30

# Experimentos
intentos = 1000

# Contadores
conteo_3_blancas_2_rojas = 0
conteo_mismo_color = 0

for i in range(intentos):
    seleccion = random.choices(caja, k=5)
    
    conteo_blancas = seleccion.count("White")
    conteo_rojas = seleccion.count("Red")
    conteo_verdes = seleccion.count("Green")
    
    if conteo_blancas == 3 and conteo_rojas == 2:
        conteo_3_blancas_2_rojas += 1
        
    elif conteo_blancas == 5 or conteo_rojas == 5 or conteo_verdes == 5:
        conteo_mismo_color += 1

# Calcular probabilidades
probabilidad_3_blancas_2_rojas = conteo_3_blancas_2_rojas / intentos
probabilidad_mismo_color = conteo_mismo_color / intentos

print(f"Probabilidad de tomar 3 blancas y 2 rojas: {np.round(probabilidad_3_blancas_2_rojas * 100, 2)}%")
print(f"Probabilidad de tomar todas del mismo color: {np.round(probabilidad_mismo_color * 100, 2)}%")
```
